PROBLEM 2 – Constituency parsing

```python
import stanza

# Download the English model
stanza.download('en')
```

```python
sentences=["Lucy plays with friends","This movie is careless and unfocused","She buys a gift with gold"]

nlp=stanza.Pipeline(lang='en', processors='tokenize,pos,constituency')#install the enlgish pipeline with tokenize, pos, and constituency parsers
doc=nlp("Lucy plays with friends")
print("Parsing of Lucy plays with friends")
for sentence in doc.sentences:
    print(sentence.constituency)
print("Parsing of This movie is careless and unfocused")
doc=nlp("This movie is careless and unfocused")
for sentence in doc.sentences:
    print(sentence.constituency)
print("Parsing of She buys a gift with gold")
doc=nlp("She buys a gift with gold")
for sentence in doc.sentences:
    print(sentence.constituency)
```

PROBLEM 3 – Reading the data

```python
import pandas as pd
```

```python
climate_change_df=pd.read_csv("climate_change.csv",delimiter=',',encoding='cp1252')#import modules
Gangs_df=pd.read_csv("Gangs.csv",delimiter=',',encoding='cp1252')
Thatcher_df=pd.read_csv("Thatcher.csv",delimiter=',',encoding='cp1252')
```

```python
climate_change_df=climate_change_df.loc[~climate_change_df.Elementary.isnull()]#locate the null cells and eliminate them
climate_change_df
```

```python
Gangs_df=Gangs_df.loc[~Gangs_df.Elementary.isnull()]#locate the null cells and eliminate them
Gangs_df
```

```python
Thatcher_df=Thatcher_df.loc[~Thatcher_df.Elementary.isnull()]#locate the null cells and eliminate them
Thatcher_df
```

```python
df_all = pd.concat([climate_change_df, Gangs_df, Thatcher_df], axis=0)#concatenate the data frames
print("Amount of rows:",len(df_all))
```

```python
#isolate and print the first rows of the elementary and advanced texts
elementary_text_0 = df_all.iloc[0]['Elementary']
advanced_text_0 = df_all.iloc[0]['Advanced']

print("Elementary Text:", elementary_text_0)
print("\nAdvanced Text:", advanced_text_0)
```

PROBLEM 4 – Analyzing the data

```python
elementary_lists = list(df_all['Elementary'])
advanced_lists = list(df_all['Advanced'])

def data_analysis(texts):
    sentence_length_array=[]
    preposition_count_array=[]
    noun_count_array=[]

    for text in texts:
      doc=nlp(text)
      preposition_count=0
      noun_phrase_count=0

      for sentence in doc.sentences:
        sentence_constituency=sentence.constituency.pretty_print()
        preposition_count+=sentence_constituency.count('PP')
        noun_phrase_count+=sentence_constituency.count('NP')

      sentence_length_array.append(len(doc.sentences))
      preposition_count_array.append(preposition_count)
      noun_count_array.append(noun_phrase_count)

    average_sentence_length=sum(sentence_length_array)/len(sentence_length_array)
    average_preposition_count=sum(preposition_count_array)/len(preposition_count_array)
    average_noun_count=sum(noun_count_array)/len(noun_count_array)

    return average_sentence_length, average_preposition_count, average_noun_count


average_sentence_num, prep_phrases_amt, noun_phrases_amt=data_analysis(elementary_lists)
print("Elementary Text information, Average Sentence Length:",average_sentence_num," Average Preposition Count:", prep_phrases_amt," Average Noun Count:", noun_phrases_amt)

average_sentence_num, prep_phrases_amt, noun_phrases_amt=data_analysis(advanced_lists)
print("Advanced Text information, Average Sentence Length:",average_sentence_num," Average Preposition Count:", prep_phrases_amt," Average Noun Count:", noun_phrases_amt)
```
